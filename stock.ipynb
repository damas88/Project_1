```python
import yfinance as yf
import datetime
import pandas as pd
import numpy as np  # Not used in the optimized code

from google.colab import files  # Assuming you need it for Colab

stock_tickers = ["0056.TW", "00878.TW", "00923.TW", "00733.TW", "006208.TW", "00919.TW", "00713.TW", "00687B.TWO", "2002.TW"]

# Download data for all tickers efficiently using list comprehension
dfs = [yf.download(ticker, start=datetime.datetime(year=2023, month=1, day=1), end=datetime.date.today(), interval="1d") for ticker in stock_tickers]

# Concatenate DataFrames with clear column names
etf_data = pd.concat([df['Close'] for df in dfs], axis=1,
                             keys=[ticker.split('.')[0] for ticker in stock_tickers])


# Resample to monthly, keeping 'Close' and rounding to 2 decimals
etf_data = etf_data.resample('M').last().round(2)

# Set index as datetime format
etf_data.index = etf_data.index.strftime('%Y-%m-%d')

# Save to CSV
etf_data.to_excel('etf_data.ods', engine='odf')

# Optionally upload to Colab storage
files.download('etf_data.ods')
```

```python
import yfinance as yf
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

stock_tickers = ["006208.TW"]
#stock_tickers = ["0056.TW","00878.TW","00923.TW","00733.TW","006208.TW","00919.TW","00713.TW","00687B.TWO"]
data_matrix = []
valid_tickers = []
for stock_id in stock_tickers:
  today = datetime.date.today()
  start_date = today - datetime.timedelta(days=365*3)  # Adjust as needed
  # Calculate the end date (last day of December in the previous year)
  current_year = datetime.datetime.now().year
  end_date = datetime.datetime(year=current_year - 1, month=12, day=31)

  #df = yf.download("0056.TW", start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'))
  df = yf.Ticker(stock_id).history(start=start_date, end=end_date)
  if df.empty:
      print(f"Data for {stock_id} is less than　3　years. Using available data.")
      df = yf.Ticker(stock_id).history(period="max") # Fetch all available data
  # Check if data is less than 1 year
  if len(df) < 365:
      print(f"Data for {stock_id} is less than 1 year. Skipping calculations.")
      continue # Skip to the next stock
  df_year = df.groupby(pd.Grouper(freq='Y')).agg({"High": "max", "Low": "min", "Close": "mean","Dividends": "sum"})

  dividend_mean = df_year['Dividends'].mean()
  yields = df_year['Dividends'] / df_year['Close']
  yield_avg = yields.mean()
  discount_rates = [0.05, 0.06, 0.07]
  price_targets = dividend_mean / np.array(discount_rates)
  new_data = [dividend_mean, yield_avg] + price_targets.tolist()
  data_matrix.append(new_data)
  valid_tickers.append(stock_id)
```

```python
ticker = ["006208.TW"]
#https://www.kaggle.com/code/asimislam/stock-performance-with-yahoo-finance-yfinance
# time period and interval
yf_period   = "max"   # 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
yf_interval = "1d"    # 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo

#  1.  Create dataframe yf_price with yf.download
yf_returns = yf.download(
        tickers = ticker,       # tickers list or string as well
        period = yf_period,      # optional, default is '1mo'
        interval = yf_interval,  # fetch data by intervaal
        group_by = 'ticker',     # group by ticker
        auto_adjust = True,      # adjust all OHLC (open-high-low-close)
        prepost = True,          # download market hours data
        threads = True,          # threads for mass downloading
        proxy = None)            # proxy


#  2.  Select 'Close' (price at market close) column only
yf_returns = yf_returns.loc[:,['Close']]
yf_returns.columns = ticker

yf_returns.tail(10)


#  4.  calculate percentage changes with "pct_change()"
#  multiply by 100 to get percentage value
#  round off the percentages to 2 decimal points
yf_returns = round(yf_returns.pct_change()*100, 2)
yf_returns.tail(10)
```
